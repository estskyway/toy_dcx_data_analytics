```python
import pymongo as mg
import pandas as pd
```

```python
client = mg.MongoClient(host='mongodb://localhost:27017')
```

```python
database = client['small_team_project']
```

```python
collection = database['hidoc_site']
```

```python
cursor = collection.find({})
```

```python
list_hidoc = list(cursor)
```

```python
list_hidoc
```

```python
df_HiDoc = pd.DataFrame(list_hidoc)
```

```python
df_HiDoc.info()
```

```python
df_HiDoc[-10:]
```

```python
columns_drop = ['_id']
df = df_HiDoc.drop(columns = columns_drop)
```

```python
df
```

```python
df_HiDoc_dropdupl = df.drop_duplicates()
```

```python
df_HiDoc_dropdupl.info()
```

```python
df_HiDoc_dropdupl
```

```python
df_HiDoc_dropduplicated = df_HiDoc_dropdupl.reset_index(drop=True)
df_HiDoc_dropduplicated
```

```python
df_HiDoc_dropduplicated.to_csv('../datasets/hidoc_database1.csv')
```
